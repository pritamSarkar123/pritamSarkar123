# Url matching

```python
urls = '''
https://www.google.com
http://coreyms.com
https://youtube.com
https://www.nasa.gov
'''

pattern = re.compile(r'https?://(www\.)?(\w+)(\.\w+)')



# matches = pattern.finditer(urls)

# for match in matches:
#     print(match.group(3))
```

# Group wise match

```python
#group zero
matches = pattern.finditer(urls)
for match in matches:
    print(match.group(0))
```

```python
#group one(here the optional one)
matches = pattern.finditer(urls)
for match in matches:
    print(match.group(1)) 
```

```python
#domain names
matches = pattern.finditer(urls)
for match in matches:
    print(match.group(2)) 
```

```python
matches = pattern.finditer(urls)
for match in matches:
    print(match.group(3)) 
```

```python
#every time the pattern matches substritute with
#\2\3 group2 and group3
subbed_urls = pattern.sub(r'\2\3', urls)

print(subbed_urls)
```

# Functions

pattern.match(text) <-checks at the the First position of the text, if not present then None 

pattern.search(text)<- checks through the entire text and return the first occurance

pattern.findall(text)<-checks through the entire text and return all occurances(groupes or all) in list format

pattern.finditer(text)<-checks through the entire text and return all occurances in iterator format,, best choice

pattern.sub(r'\i\j\.....\z',text)<- checks the entire text, find the patterns and substritute them with group i group j.... group z
