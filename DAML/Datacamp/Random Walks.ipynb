```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(123)
```

```python
all_walks=[]
for y in range(10000):
    random_walk=[0]
    for x in range(100):
        step=random_walk[-1]
        dice=np.random.randint(1,7)
        if dice<=2:
            step=max(0,step-1)
        elif dice<=5:
            step+=1
        else:
            step+=np.random.randint(1,7)
        #falling down probability
        if np.random.rand()<=0.001:
            step=0
        random_walk.append(step)
    all_walks.append(random_walk)
```

```python
np_aw=np.array(all_walks)
```

```python
np_aw.shape
```

```python
plt.plot(np_aw)
plt.show()
```

```python
np_aw_t=np.transpose(np_aw)
plt.plot(np_aw_t)
plt.show()
```

```python
walk_ends=np_aw_t[-1,:]
plt.hist(walk_ends,bins=10)
plt.xticks([x for x in range(0,150,10)],[x for x in range(0,150,10)])
plt.grid(True)
plt.show()
```

```python
count_atleast_60=0
for e in walk_ends:
    if e>=60:
        count_atleast_60+=1
print(f'probability of reaching atleast 60 steps in 10000 trials is {100*(count_atleast_60/len(walk_ends))}%')
```
